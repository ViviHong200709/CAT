```python
import matplotlib.pyplot as plt
import numpy as np

def str2arr(a):
    a=a.split("	")
    return [float(i) for i in a]

def draw(Y1,Y2,Y3,standard,dataset,stg="MFI",xlabel="question nums",Y4=None):
    max_len = 20
    X=np.linspace(1,max_len,max_len)
    Y1 = str2arr(Y1)
    Y2 = str2arr(Y2)
    Y3 = str2arr(Y3)
    if Y4:
        Y4 = str2arr(Y4)
        

    plt.figure(figsize=(8,6))  # 定义图的大小
    plt.xlabel(xlabel)
    plt.ylabel(standard)
    plt.title(dataset)

    plt.plot(X,Y1[1:max_len+1],label='Random')
    plt.plot(X,Y2[1:max_len+1],label=f'vanilla {stg}')
    plt.plot(X,Y3[1:max_len+1],label=f'distill-balltree {stg}')
    if Y4:
        plt.plot(X,Y4[1:max_len+1],label=f'diversity-balltree {stg}')
    plt.legend(loc='lower right')
    plt.show()
```
